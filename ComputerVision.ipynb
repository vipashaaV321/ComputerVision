Create Black Image

```
import numpy as np
blackimg=np.zeros(shape=(512,512,3),dtype=np.int16)
```

```
blackimg.shape
```

```
pyplot.imshow(blackimg)
```

#Draw Shapes

#Circle

```
cirimg=cv2.circle(img=blackimg,center=(400,100),radius=50,color=(255,0,0),thickness=8)
```

```
pyplot.imshow(cirimg)
```

```
cirimg=cv2.circle(img=blackimg,center=(100,100),radius=50,color=(255,155,0),thickness=-1)
pyplot.imshow(cirimg)
```

#Rectangle

```
cv2.rectangle(blackimg,pt1=(200,200),pt2=(300,300),color=(251,45,67),thickness=5)
pyplot.imshow(blackimg)
```

#Triangle

```
vertices=np.array([[100,400],[310,350],[400,450]],np.int32)
plts=vertices.reshape(-1,1,2)
cv2.polylines(blackimg,[plts],isClosed=True,color=(0,7,255),thickness=3)
pyplot.imshow(blackimg)
```

```
vertices=np.array([[100,400],[310,350],[400,450]],np.int32)
plts=vertices.reshape(-1,1,2)
cv2.fillPoly(blackimg,[plts],color=(201,7,255))
pyplot.imshow(blackimg)
```

```
cv2.line(blackimg,pt1=(512,0),pt2=(0,512),color=(255,0,25),thickness=3)
pyplot.imshow(blackimg)
```

```

# Window name in which image is displayed 
window_name = 'Image'
  
# font 
font = cv2.FONT_HERSHEY_SIMPLEX 
  
# org 
org = (40, 40) 
  
# fontScale 
fontScale = 1
   
# Blue color in BGR 
color = (235, 166, 3) 
  
# Line thickness of 2 px 
thickness = 2
   
# Using cv2.putText() method 
image = cv2.putText(blackimg, 'Vipashaa', org, font,  
                   fontScale, color, thickness, cv2.LINE_AA) 
   
# Displaying the image 
pyplot.imshow(image)  
```
